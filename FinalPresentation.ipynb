### Finding Pi
- Uses Monte Carlo method.
    - A square of a definable area centered at (0,0) is created.
    - A large number of points are plotted inside the square at random.
    - If the point is recorded to be within a circle with diameter equal to the width of the square, it adds 1 to a counting value.
    - The number of points inside the circle divided by the total number of points plotted approximately equals the area.

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

#Define a square region with sides of a certain length. In that region define a circle with a radius of your choosing.
#Monte Carlo Technique

def area(x):
    #bounds of the square in both x and y directions
    a = 0
    b = x
    #circle of radius 5, centered at (0,0)

    n = 100000
    pointsinside = 0
    for i in range(n):
        point = (rand.uniform(a,b),rand.uniform(a,b))
        if point[0]**2 + point[1]**2 <= x**2:
            pointsinside += 1
    #times 4 since the region is a quarter circle
    area = (pointsinside/n)*4

    return area
print(area(1))
```

- Does not return the same value consistently
- For more accuracy, n is increased. However, high n values increase the run time of the code significantly.
